````python
import numpy as np

# Define the sample data
sample = np.array([12, 7, 9, 15, 8, 13, 10, 11, 14, 10])  # Replace with your own sample data

# Define the bootstrap function to compute a confidence interval
def bootstrap_ci(sample, stat_func, n_bootstraps=1000, ci=95):
    """
    Compute the bootstrap confidence interval for a given population parameter.

    Parameters:
        sample (numpy array): The sample data.
        stat_func (function): The statistical function to apply on the sample (e.g., np.mean, np.median).
        n_bootstraps (int): Number of bootstrap samples to generate.
        ci (float): Confidence interval percentage.

    Returns:
        tuple: Lower and upper bounds of the bootstrap confidence interval.
    """
    # Generate bootstrap samples and compute the statistic for each sample
    bootstrapped_stats = np.array([stat_func(np.random.choice(sample, size=len(sample), replace=True))
                                   for _ in range(n_bootstraps)])
    
    # Calculate the lower and upper percentiles for the confidence interval
    lower_bound = np.percentile(bootstrapped_stats, (100 - ci) / 2)
    upper_bound = np.percentile(bootstrapped_stats, 100 - (100 - ci) / 2)
    
    return lower_bound, upper_bound

# Calculate the 95% bootstrap confidence interval for the population mean
mean_ci = bootstrap_ci(sample, np.mean)

# Print the result
print(f"95% Bootstrap Confidence Interval for the Population Mean: {mean_ci}")

# COMMENT: To calculate a 95% bootstrap confidence interval for a different population parameter,
# such as the median, simply replace `np.mean` with `np.median`:
# median_ci = bootstrap_ci(sample, np.median)
# print(f"95% Bootstrap Confidence Interval for the Population Median: {median_ci}")
````

We'll start by creating a visualization that compares the initial and final health scores of each patient to get an idea of how the scores have changed.

Quantitative Analysis:
    Analysis Steps:
        Calculate the Difference: Compute the difference between the initial and final health scores for each patient.
        Bootstrap Sampling: Use bootstrapping to resample these differences and calculate the mean difference across many resamples.
        Construct Confidence Interval: Construct a confidence interval for the mean difference.
        Hypothesis Testing: Check if the confidence interval includes zero. If it does, we fail to reject the null hypothesis of "no effect." If it doesn't, we reject the null hypothesis and conclude that the vaccine had a statistically significant effect.
    Code:

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Use the matplotlib inline magic command to render images in PNG format in JupyterHub
%matplotlib inline

# Set random seed for reproducibility
np.random.seed(42)

# URL of the CSV file
csv_url = "https://raw.githubusercontent.com/mikuknightforever/STA130/refs/heads/main/Ali_tech.csv"

# Read the CSV file directly from the URL
df = pd.read_csv(csv_url)

# Calculate the difference between Final and Initial Health Scores
df['HealthScoreDifference'] = df['FinalHealthScore'] - df['InitialHealthScore']

# Plot the initial and final health scores for each patient
plt.figure(figsize=(10, 6))
plt.plot(df['PatientID'], df['InitialHealthScore'], marker='o', label='Initial Health Score', linestyle='--')
plt.plot(df['PatientID'], df['FinalHealthScore'], marker='o', label='Final Health Score', linestyle='-')
plt.xlabel('Patient ID')
plt.ylabel('Health Score')
plt.title('Initial vs. Final Health Scores of Patients')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Perform bootstrapping to estimate the mean difference
def bootstrap_mean_diff(data, n_bootstrap=10000):
    """Bootstrap the mean of the differences."""
    boot_means = []
    for _ in range(n_bootstrap):
        # Resample with replacement
        sample = np.random.choice(data, size=len(data), replace=True)
        # Calculate the mean of the resampled data
        boot_means.append(np.mean(sample))
    return np.array(boot_means)

# Bootstrap the health score differences
bootstrapped_means = bootstrap_mean_diff(df['HealthScoreDifference'].values)

# Calculate 95% confidence interval
confidence_interval = np.percentile(bootstrapped_means, [2.5, 97.5])
print(f"95% Confidence Interval for the Mean Health Score Difference: {confidence_interval}")

# Plot the bootstrapped distribution of the mean health score differences
plt.figure(figsize=(10, 6))
plt.hist(bootstrapped_means, bins=30, color='skyblue', edgecolor='black')
plt.axvline(confidence_interval[0], color='red', linestyle='dashed', linewidth=2, label='95% CI Lower Bound')
plt.axvline(confidence_interval[1], color='red', linestyle='dashed', linewidth=2, label='95% CI Upper Bound')
plt.title('Bootstrapped Distribution of Mean Health Score Differences')
plt.xlabel('Mean Health Score Difference')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

# Show the histogram plot
plt.show()
````
